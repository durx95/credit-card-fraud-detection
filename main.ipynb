```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report , confusion_matrix ,  roc_auc_score
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

### visualize class imbalance

```python
sns.countplot(x='Class',data=df)
plt.title('class distribution')
plt.show()
```

### separate features and target


```python
x =df.drop('Class',axis=1)
y = df['Class']
```

### train-test split

```python
x_train , x_test ,y_train ,y_test = train_test_split(
    x , y , test_size=0.2 , random_state=42 , stratify=y
)
```

### feature scaling

```python
scaler = StandardScaler()

x_train['Amount'] = scaler.fit_transform(x_train[['Amount']])
x_test['Amount'] = scaler.transform(x_test[["Amount"]])

x_train['Time'] = scaler.fit_transform(x_train[["Time"]])
x_test['Time'] = scaler.transform(x_test[["Time"]])
```

```python
model = LogisticRegression(class_weight={0:1,1:20},max_iter=1000)
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
```

### evaluate model

```python
print(confusion_matrix(y_test,y_pred))

print(classification_report(y_test,y_pred))

roc_auc = roc_auc_score(y_test,model.predict_proba(x_test)[:,1])

print("roc-auc:",roc_auc)
```

### find the probability.

```python
y_prob = model.predict_proba(x_test)[:,1]
```

### apply the custom threshold


```python
threshold =0.83
y_pred_new = (y_prob > threshold).astype(int)
```

### evaluate again

```python
print("confusion matrix:\n",confusion_matrix(y_test,y_pred_new))

print("classification report:\n",classification_report(y_test,y_pred_new))
```

### save the model

```python
import joblib
joblib.dump(model, "fraud_model.pkl")
```
